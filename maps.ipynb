```python
import ee
import geemap

# Zert test 
ee.Initialize()

# Define a region of interest (ROI) ,  ; , ; , ; , 
roi = ee.Geometry.Polygon([[
    [-90.545164,32.667379], 
    [-90.533051, 32.667207], 
    [-90.529994, 32.655962], 
    [-90.542514, 32.655673]
]])

# Define a time period
start_date = '2022-02-01'
end_date = '2022-03-11'

# Load MODIS/Terra Vegetation Indices collection (MOD13Q1 provides 16-day NDVI)
modis_vi_collection = ee.ImageCollection('MODIS/006/MOD13Q1') \
    .filterDate(start_date, end_date) \
    .filterBounds(roi) \
    .select('NDVI')

# Function to scale MODIS NDVI values and reproject to EPSG:4326
def scale_and_reproject(image):
    return image.multiply(0.0001).reproject('EPSG:4326', None, 250).copyProperties(image, ['system:time_start'])

# Scale and reproject the NDVI values
scaled_vi_collection = modis_vi_collection.map(scale_and_reproject)

# Visualization parameters for NDVI
ndvi_vis_params = {
    'min': -0.1,
    'max': 0.8,
    'palette': ['blue', 'white', 'green']
}

# Create an interactive map
m = geemap.Map(center=[32.658312, -90.536774], zoom=8)
m.add_basemap('HYBRID')

# Add the NDVI ImageCollection to the map with a time slider
m.add_time_slider(ee_object=scaled_vi_collection, vis_params=ndvi_vis_params, region=roi, date_format='YYYY-MM-dd')

# Add layer control and display the map
m.addLayerControl()
m
```
